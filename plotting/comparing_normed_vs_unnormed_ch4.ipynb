```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### This notebook compares transforms with their normed versions (normed s.t. tr(r=a)=a )
# N_train = N_test = 1000, a = 4.5, Emax = 400


Path = '/home/ach/radial-physics_new/radial-physics/output/GPR_old_cweight/normed_old_TR'
Path_unnormed = '/home/ach/radial-physics_new/radial-physics/output/GPR_old_cweight/dataset_comp_old_calcs'

no_of_tr = [2,3,4,7,8]
# minimia_all_tr_qm9 = []
minimia_all_tr_ch4_normed = []
minimia_all_tr_ch4_unnormed = []
# minimia_all_tr_gold = []

for i in no_of_tr:
    
    #Define/reset lists + open data
    # factor_qm9 = [] #QM9
    # rmse_qm9 = []
    # data_qm9 = open(Path+f'/Data_TR{i}_qm9','r')

    # factor_gold = [] #Gold
    # rmse_gold = []
    # data_gold = open(Path+f'/Data_TR{i}_gold','r')

    factor_ch4_normed = [] #Methane
    rmse_ch4_normed = []
    data_ch4_normed = open(Path+f'/Data_TR{i}000_random-ch4-10k','r')

    factor_ch4_unnormed = [] #Methane
    rmse_ch4_unnormed = []
    data_ch4_unnormed = open(Path_unnormed+f'/Data_TR{i}_random-ch4-10k','r')

    # for row in data_qm9: #QM9
    #     row = row.split(' ')
    #     factor_qm9.append(float(row[0]))
    #     rmse_qm9.append(float(row[1]))
    
    # for row in data_gold: #Gold
    #     row = row.split(' ')
    #     factor_gold.append(float(row[0]))
    #     rmse_gold.append(float(row[1]))
    
    for row in data_ch4_normed: #Methane
        row = row.split(' ')
        factor_ch4_normed.append(float(row[0]))
        rmse_ch4_normed.append(float(row[1]))

    for row in data_ch4_unnormed: #Methane
        row = row.split(' ')
        factor_ch4_unnormed.append(float(row[0]))
        rmse_ch4_unnormed.append(float(row[1]))
    
    # print('factors: ', factor_ch4)
    # print('rmse: ', rmse_ch4)
        
    #indices of minima and minima (for plot later)
    # minimum_indx_qm9 = np.argmin(rmse_qm9,axis=0)
    minimum_indx_ch4_normed = np.argmin(rmse_ch4_normed,axis=0)
    minimum_indx_ch4_unnormed = np.argmin(rmse_ch4_unnormed,axis=0)
    # minimum_indx_gold = np.argmin(rmse_gold,axis=0)
    
    # minimia_all_tr_qm9.append(rmse_qm9[minimum_indx_qm9])
    minimia_all_tr_ch4_normed.append(rmse_ch4_normed[minimum_indx_ch4_normed])
    minimia_all_tr_ch4_unnormed.append(rmse_ch4_unnormed[minimum_indx_ch4_unnormed])
    # minimia_all_tr_gold.append(rmse_gold[minimum_indx_gold])

    plt.plot(factor_ch4_normed, rmse_ch4_normed , "b-o", label="CH$_{4}$")
    plt.plot(factor_ch4_unnormed, rmse_ch4_unnormed, "r-o", label="CH$_{4}$ unnormed")
    plt.xlabel("Factor")
    plt.ylabel("RMSE [kcal/mol]")
    plt.title(f"Transform {i}")
    # plt.legend(bbox_to_anchor=(1, 1.), ncol=5)
    plt.legend()
    plt.show()
    
```

```python
### one plot for each data set with all minima:

plt.plot(no_of_tr, minimia_all_tr_ch4_normed, "b:o", label="CH$_{4}$ normed")
plt.plot(no_of_tr, minimia_all_tr_ch4_unnormed, "r:o", label="CH$_{4}$ unnormed")
plt.xlabel("Transform")
plt.ylabel("RMSE [kcal/mol]")
plt.title("Minimum RMSE for all transforms")
plt.legend()
plt.show()
```
